```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
file_path = "Visadataset_cleaned.csv"  # Update if needed
df = pd.read_csv(file_path)
```

```python
#Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns
```

```python
#Apply Label Encoding to each categorical column
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Save encoder if needed later
```

```python
#Save the fully numeric dataset
df.to_csv("Visadataset_numerical.csv", index=False)
```

```python
# Show label encoding mapping for each column
for col, le in label_encoders.items():
    print(f"\nLabel encoding for column '{col}':")
    for original, encoded in zip(le.classes_, le.transform(le.classes_)):
        print(f"  {original} â†’ {encoded}")
```
