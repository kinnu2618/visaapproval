```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer
```

```python
file_path = "Visadataset_numerical.csv"  # Update if needed
data = pd.read_csv(file_path)
```

```python
# Standardization
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled
```

```python
# Min-Max Scaling
min_max_scaler = MinMaxScaler()
data_normalized = min_max_scaler.fit_transform(data)
```

```python
# L2 Normalization
normalizer = Normalizer()
data_l2_normalized = normalizer.fit_transform(data)
```

```python
print("Standardized Data:\n", data_scaled)
print("Min-Max Scaled Data:\n", data_normalized)
print("L2 Normalized Data:\n", data_l2_normalized)
```
