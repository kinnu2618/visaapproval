```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Visadataset.csv')
```

```python
df.head()
```

```python
# Set style
sns.set(style="whitegrid")
```

* Distribution of case status (Certified/Denied)

* Education level vs case status

* Continent distribution of applicants

* Job experience impact on case status

* Regional distribution of employment

* Full-time position impact

* Wage distribution analysis

* Company size analysis

* Year of establishment distribution

* Job training requirement impact

```python
# 1. Case Status Distribution
plt.figure(figsize=(8,6))
sns.countplot(x='case_status', data=df)
plt.title('Distribution of Visa Case Status')
plt.xlabel('Case Status')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Education Level vs Case Status
plt.figure(figsize=(10,6))
sns.countplot(x='education_of_employee', hue='case_status', data=df)
plt.title('Education Level vs Visa Case Status')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Case Status')
plt.show()
```

```python
# 3. Continent Distribution
plt.figure(figsize=(10,6))
sns.countplot(y='continent', data=df, order=df['continent'].value_counts().index)
plt.title('Distribution of Applications by Continent')
plt.xlabel('Count')
plt.ylabel('Continent')
plt.show()
```

```python
# 4. Job Experience vs Case Status
plt.figure(figsize=(8,6))
sns.countplot(x='has_job_experience', hue='case_status', data=df)
plt.title('Job Experience vs Visa Case Status')
plt.xlabel('Has Job Experience')
plt.ylabel('Count')
plt.show()
```

```python
# 5. Region of Employment Distribution
plt.figure(figsize=(10,6))
sns.countplot(y='region_of_employment', data=df, order=df['region_of_employment'].value_counts().index)
plt.title('Distribution of Applications by Region of Employment')
plt.xlabel('Count')
plt.ylabel('Region of Employment')
plt.show()
```

```python
# 6. Full Time Position vs Case Status
plt.figure(figsize=(8,6))
sns.countplot(x='full_time_position', hue='case_status', data=df)
plt.title('Full Time Position vs Visa Case Status')
plt.xlabel('Full Time Position')
plt.ylabel('Count')
plt.show()
```

```python
# 7. Prevailing Wage Distribution by Case Status
plt.figure(figsize=(10,6))
sns.boxplot(x='case_status', y='prevailing_wage', data=df[df['unit_of_wage'] == 'Year'])
plt.title('Prevailing Wage Distribution by Case Status (Yearly)')
plt.xlabel('Case Status')
plt.ylabel('Prevailing Wage (Yearly)')
plt.yscale('log')  # Using log scale due to wide wage range
plt.show()
```

```python
# 8. Company Size (no_of_employees) vs Case Status
plt.figure(figsize=(10,6))
sns.boxplot(x='case_status', y='no_of_employees', data=df)
plt.title('Company Size vs Visa Case Status')
plt.xlabel('Case Status')
plt.ylabel('Number of Employees')
plt.yscale('log')  # Using log scale due to wide range
plt.show()
```

```python
# 9. Year of Establishment Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['yr_of_estab'], bins=30)
plt.title('Distribution of Year of Establishment for Companies')
plt.xlabel('Year of Establishment')
plt.ylabel('Count')
plt.show()
```

```python
# 10. Job Training Requirement vs Case Status
plt.figure(figsize=(8,6))
sns.countplot(x='requires_job_training', hue='case_status', data=df)
plt.title('Job Training Requirement vs Visa Case Status')
plt.xlabel('Requires Job Training')
plt.ylabel('Count')
plt.show()
```
